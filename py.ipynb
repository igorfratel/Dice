```python
#!/usr/bin/python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import random

def criaGrafico(n):
    plt.figure(figsize=(20, 10), dpi=80)
    plt.title('Grafico de Frequencia Relativa')
    plt.xlabel('Lancamentos')
    plt.ylabel('Frequencia Relativa')
    plt.xlim(0, n)
    plt.ylim(0, 1)
    red_patch = mpatches.Patch(color='red', label='Numero 1')
    green_patch = mpatches.Patch(color='green', label='Numero 2')
    blue_patch = mpatches.Patch(color='blue', label='Numero 3')
    yellow_patch = mpatches.Patch(color='yellow', label='Numero 4')
    black_patch = mpatches.Patch(color='black', label='Numero 5')
    magenta_patch = mpatches.Patch(color='magenta', label='Numero 6')
    plt.legend(handles=[red_patch, green_patch, blue_patch, yellow_patch, black_patch, magenta_patch])

def main():
    n = 0
    i = 1
    x = 0
    um = 0.0
    dois = 0.0
    tres = 0.0
    quatro = 0.0
    cinco = 0.0
    seis = 0.0
    v1 = []
    v2 = []
    v3 = []
    v4 = []
    v5 = []
    v6 = []
    n = int(input("Insira quantas vezes o dado sera lancado: "))
    criaGrafico(n)
    while i<=n+1:
        x = random.randint(1, 6)
        if x == 1:
            um +=  1

        elif x == 2:
            dois += 1

        elif x == 3:
            tres += 1

        elif x == 4:
            quatro += 1

        elif x == 5:
            cinco += 1

        else:
            seis += 1
        v1.append([i, um/i])
        v2.append([i, dois/i])
        v3.append([i, tres/i])
        v4.append([i, quatro/i])
        v5.append([i, cinco/i])
        v6.append([i, seis/i])
        i += 1
    plt.plot(v1, color = 'r')
    plt.plot(v2, color = 'g')
    plt.plot(v3, color = 'b')
    plt.plot(v4, color = 'y')
    plt.plot(v5, color = 'black')
    plt.plot(v6, color = 'm')
    plt.grid()
    plt.show()

if __name__ == "__main__":
    main()
    
```
